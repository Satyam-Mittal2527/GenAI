```python
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings
```

```python
from langchain_classic.schema import Document
doc1 = Document(
        page_content="""
Virat Kohli is one of the greatest batters in IPL history. He has spent his entire IPL career with Royal Challengers Bengaluru (RCB). Known for his aggressive batting style and consistency, Kohli has scored over 8000 IPL runs and has won multiple Orange Caps. He was the captain of RCB for several seasons and remains one of the most popular cricketers in the world.
""",
        metadata={
            "player": "Virat Kohli",
            "team": "RCB",
            "role": "Batsman",
            "country": "India",
            "id": 1
        }
    ),

doc2 = Document(
        page_content="""
MS Dhoni is the legendary captain of Chennai Super Kings (CSK). He is regarded as one of the best finishers in cricket history and has led CSK to multiple IPL championships. Dhoni is famous for his calm leadership, quick wicketkeeping skills, and ability to finish matches under pressure.
""",
        metadata={
            "player": "MS Dhoni",
            "team": "CSK",
            "role": "Wicketkeeper Batter",
            "country": "India",
            "id": 2
        }
    ),

doc3 = Document(
        page_content="""
Rohit Sharma is one of the most successful IPL captains, having led Mumbai Indians (MI) to multiple IPL titles. He is an elegant opening batter known for his timing and ability to score big hundreds. Rohit has represented Mumbai Indians throughout most of his IPL career.
""",
        metadata={
            "player": "Rohit Sharma",
            "team": "MI",
            "role": "Opening Batter",
            "country": "India",
            "id": 3
        }
    ),

doc4 = Document(
        page_content="""
Jasprit Bumrah is the premier fast bowler for Mumbai Indians (MI). He is famous for his unique bowling action, deadly yorkers, and exceptional death-over bowling. Bumrah has consistently been among the top wicket-takers in the IPL and is considered one of the best fast bowlers in the world.
""",
        metadata={
            "player": "Jasprit Bumrah",
            "team": "MI",
            "role": "Fast Bowler",
            "country": "India",
            "id": 4
        }
    ),

doc5 = Document(
        page_content="""
Andre Russell is a powerful all-rounder who has been a key player for Kolkata Knight Riders (KKR). He is known for his explosive batting, massive sixes, and ability to bowl fast in crucial situations. Russell has won numerous Player of the Match awards due to his match-winning performances.
""",
        metadata={
            "player": "Andre Russell",
            "team": "KKR",
            "role": "All-rounder",
            "country": "West Indies",
            "id": 5
        }
    )
```

```python
docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
vector_store = Chroma(
    embedding_function = OpenAIEmbeddings(),
    persist_directory = 'my_chroma_db',
    collection_name = 'sample'
)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=['embeddings','documents','metadata'])
```

```python
vector_store.similarity_search(
    query = "who among these are a bowler?",
    k=1
)
```

```python
##Less store means more similar as the score represents the distance
vector_store.similarity_search_with_score(
    query = "who among these are a bowler?",
    k=1
)
```

```python
##Less store means more similar as the score represents the distance
vector_store.similarity_search_with_score(
    query = "",
    filter = {"team":"Chennai Super Kings"}
)
```
